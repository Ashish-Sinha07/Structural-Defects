```python
import google.generativeai as genai
from PIL import Image
import os
```

```python
defect_image = Image.open('images.jfif')
```

```python
defect_image
```

```python
gemini_api_key = os.getenv('GEMINI-API-KEY-3')
genai.configure(api_key=gemini_api_key)
model = genai.GenerativeModel('gemini-2.5-flash-lite')
```

```python
prompt= " Identify the structure defect in the image and suggest me how i can fix the construction issue"
```

```python
response = model.generate_content([prompt,defect_image],
                       generation_config={'temperature':0.7})
```

```python
print(response.text)
```
